# Exploratory Data Analysis (EDA)

This notebook performs exploratory data analysis on the dataset to uncover insights, patterns, and relationships.

## 1. Loading Data


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set aesthetic parameters in one step
sns.set(style="whitegrid")

# Load the dataset
df = pd.read_csv('../data/processed/cleaned_data_20240621_143909.csv')

# Display first few rows of the dataset
df.head()
```

## 2. Basic Data Exploration

### 2.1 Shape of the Dataset


```python
# Shape of the dataset
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

### 2.2 Data Types and Missing Values


```python
# Data types and missing values
print(df.info())

# Summary of missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

## 3. Data Visualization

### 3.1 Univariate Analysis


```python
# Univariate analysis: Distribution of numerical features
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plotting the distribution of numerical features
df[num_cols].hist(figsize=(15, 10), bins=20)
plt.suptitle('Distribution of Numerical Features')
plt.show()
```

### 3.2 Categorical Features


```python
# Categorical features
cat_cols = df.select_dtypes(include=['object', 'bool']).columns

# Plotting the count of categorical features
fig, axes = plt.subplots(len(cat_cols), 1, figsize=(10, 5*len(cat_cols)))

for i, col in enumerate(cat_cols):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Count of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### 3.3 Bivariate Analysis


```python
# Bivariate analysis: Numerical features vs Target
target = 'target'

fig, axes = plt.subplots(len(num_cols), 1, figsize=(10, 5*len(num_cols)))

for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x=target, y=col, ax=axes[i])
    axes[i].set_title(f'{col} vs {target}')
    axes[i].set_xlabel(target)
    axes[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```

### 3.4 Correlation Analysis


```python
# Correlation matrix
# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix
corr = numeric_df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

```

### Key Findings from EDA

1. **Correlated Features:**

2. **Significant Outliers:**
   - **Loan Amount:**
     - **Insight**: There are significant outliers in the `loan_amount` feature, especially on the higher end.
     - **Relevance**: These outliers might represent loans with a higher risk of default and should be carefully considered in the model to avoid skewing predictions.

   - **Salary:**
     - **Insight**: A few individuals have extremely high salaries compared to the rest.
     - **Relevance**: These high salaries may not necessarily correlate with a lower risk of default and might need transformation.

3. **Distribution Patterns:**
   - **Age:**
     - **Insight**: The age distribution is normal, with most individuals in the 40 to 50 age range.
     - **Relevance**: Age-related patterns can help segment borrowers into different risk categories, improving the model’s accuracy.

   - **Interest Rate:**
     - **Insight**: Interest rates have a tight distribution with few deviations.
     - **Relevance**: Consistent interest rates across loans suggest that other factors may play a more significant role in predicting defaults.
